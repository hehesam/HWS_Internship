# Trying out the model

```python
cp='79999_iter.pth'
data_path = './data'

n_classes = 19
net = BiSeNet(n_classes=n_classes)

net.cuda()
save_pth = osp.join(data_path, cp)
net.load_state_dict(torch.load(save_pth))
net.eval()

to_tensor = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225)),
])
```

```python
cap = cv2.VideoCapture(1)
```

```python
ret, frame = cap.read()
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
img = Image.fromarray(frame)
```

```python
image = img.resize((512, 512), Image.BILINEAR)
```

```python
image 
```

```python
img = to_tensor(image)
img = torch.unsqueeze(img, 0)
img = img.cuda()
```

```python
out = net(img)[0]
parsing = out.detach().squeeze(0).cpu().numpy().argmax(0)
parsing = np.array(parsing)
```

```python
index = np.where(np.bitwise_or(parsing==0, parsing==16))
```

```python
pic = np.array(image.getdata()).reshape((512,512,3))
```

```python
for id1, id2 in zip(index[0], index[1]):
    pic[id1,id2,:]=0
```

```python
img_out = Image.fromarray(pic.astype('uint8'))
```

```python
img_out
```

# Try Using opencv 

```python
import cv2 
import matplotlib.pyplot as plt
```

```python
onnx_model_path = './net.pk'
network = cv2.dnn.readNetFromONNX(onnx_model_path)
```

```python
cap = cv2.VideoCapture(1)
```

```python
ret, frame = cap.read()
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
img = Image.fromarray(frame)
image = img.resize((512, 512), Image.BILINEAR)
```

```python
image
```

```python
img_np = np.array(image.getdata()).reshape(512,512,3)
```

```python
img_np = img_np.astype('float32')/255
```

```python
img_np[:,:,0] = (img_np[:,:,0] - 0.485)/0.229
img_np[:,:,1] = (img_np[:,:,1] - 0.456)/0.224
img_np[:,:,2] = (img_np[:,:,2] - 0.406)/0.225
```

```python
img_np = img_np.transpose((2,0,1))
```

```python
img_np = np.expand_dims(img_np, axis=0)
```

```python
network.setInput(img_np)
out = network.forward()[0]
```

```python
parsing = out.argmax(0)
parsing = np.array(parsing)
```

```python
index = np.where(np.bitwise_or(parsing==0, parsing==16))
pic = np.array(image.getdata()).reshape((512,512,3))
for id1, id2 in zip(index[0], index[1]):
    pic[id1,id2,:]=0
```

```python
img_out = Image.fromarray(pic.astype('uint8'))
```
